# **Customer Behavior & Demographics Analysis**


```python
q11 = '''
SELECT str.state,
'$' || ROUND(SUM(ort.quantity*(ort.list_price*(1-ort.discount))) / COUNT(DISTINCT ort.order_id),2)
AS average_revenue_per_order
FROM stores str 
LEFT JOIN orders o 
ON o.store_id=str.store_id 
LEFT JOIN order_items ort
ON ort.order_id=o.order_id GROUP BY str.state ORDER BY average_revenue_per_order DESC;
'''

df11 = pd.read_sql_query(q11, db) 
df11
```

Identifies the average revenue a customer would make per order (total revenue/total orders) grouped by state.
As we can see, Texans tend to spend the most per order, so online services could be informed to focus on Texas markets.

```python
q12 = '''
SELECT status, COUNT(status) AS 'Amount'  FROM
(
SELECT cu.first_name,cu.last_name,
CASE WHEN COUNT(DISTINCT o.order_id)>1 THEN 'returning'
ELSE 'new'
END AS 'status'
FROM customers cu 
LEFT JOIN orders o ON 
o.customer_id=cu.customer_id
GROUP BY cu.customer_id
)
GROUP BY status;
'''

df12 = pd.read_sql_query(q12, db) 
df12
```

Identifies how many customers are new and how many are returning via a subquery for all three stores. 

```python
qx = '''
SELECT 
  store_name,
  COUNT(CASE WHEN status = 'returning' THEN 1 END) AS returning_customers,
  COUNT(CASE WHEN status = 'new' THEN 1 END) AS new_customers,
  ROUND(
    100.0 * COUNT(CASE WHEN status = 'returning' THEN 1 END) 
    / COUNT(*), 
    2
  ) || '%' AS returning_percentage
FROM (
  SELECT 
    cu.customer_id,
    s.store_name,
    CASE 
      WHEN COUNT(DISTINCT o.order_id) > 1 THEN 'returning'
      ELSE 'new'
    END AS status
  FROM customers cu
  LEFT JOIN orders o ON o.customer_id = cu.customer_id
  LEFT JOIN stores s ON s.store_id = o.store_id
  GROUP BY cu.customer_id, s.store_name
) customer_status_by_store
GROUP BY store_name ORDER BY 
ROUND(
    100.0 * COUNT(CASE WHEN status = 'returning' THEN 1 END) 
    / COUNT(*), 
    2
  ) DESC
'''

dfx = pd.read_sql_query(qx, db) 
dfx
```

Although Baldwin Bikes have had the most returning customers, Santa Cruz Bikes has the largest retention rate of customers, with 18.31% of all of its customers returning.

# **Sales & Revenue Analysis**


```python
q15 = '''
SELECT b.brand_name, '$' ||
ROUND(((SUM(ort.quantity*(ort.list_price*(1-ort.discount))))/1000),1) || ' thousand'
AS revenue
FROM brands b 
LEFT JOIN products p 
ON p.brand_id=b.brand_id
LEFT JOIN order_items ort 
ON ort.product_id=p.product_id 
GROUP BY b.brand_name 
ORDER BY 
ROUND(((SUM(ort.quantity*(ort.list_price-(ort.discount*ort.list_price))))/1000),2) DESC;



'''

df15 = pd.read_sql_query(q15, db) 
df15
```

The Trek bike brand is doing the best in terms of generating revenue across all three stores.

```python
qy = '''

WITH brand_revenue AS (
  SELECT 
    s.store_name AS store,
    b.brand_name,
    SUM(ort.quantity * (ort.list_price * (1 - ort.discount))) AS total_revenue
  FROM brands b
  JOIN products p ON p.brand_id = b.brand_id
  JOIN order_items ort ON ort.product_id = p.product_id
  JOIN orders o ON o.order_id = ort.order_id
  JOIN stores s ON s.store_id = o.store_id
  GROUP BY s.store_name, b.brand_name
)
SELECT store, brand_name,
  '$' || ROUND(total_revenue / 1000, 1) || ' thousand' AS revenue
FROM brand_revenue br
WHERE total_revenue = (
  SELECT MAX(total_revenue)
  FROM brand_revenue
  WHERE store = br.store
)
ORDER BY store;
'''

dfy = pd.read_sql_query(qy, db) 
dfy
```

Identifies the best-selling brand by store.

```python
q16= '''
SELECT p.product_name, SUM(ort.quantity) AS Total_quantity_ordered, '$' ||
ROUND(((SUM(ort.quantity*(ort.list_price*(1-ort.discount))))),2) AS total_revenue
FROM products p
LEFT JOIN order_items ort ON ort.product_id=p.product_id
GROUP BY p.product_name ORDER BY SUM(ort.quantity) DESC,
ROUND(((SUM(ort.quantity*(ort.list_price*(1-ort.discount))))),2) DESC LIMIT 10;

'''
df16 = pd.read_sql_query(q16, db) 
df16
```

The Trek Slash 8 27.5 has made the most total revenue overall with 555k in sales across all 3 stores, but it doesn't make top the list in terms of quantity ordered, with the Electra Cruiser 1 (24-inch) topping the list with 296 sales and 72k in total revenue.

```python
q17 = '''
SELECT cat.category_name, 
'$' ||
ROUND(((SUM(ort.quantity*(ort.list_price*(1-ort.discount))))/1000),2) || ' thousand'
AS total_revenue
FROM categories cat 
LEFT JOIN products p 
ON p.category_id=cat.category_id
LEFT JOIN order_items ort ON ort.product_id=p.product_id
GROUP BY cat.category_name
ORDER BY ((SUM(ort.quantity*(ort.list_price*(1-ort.discount))))/1000) DESC;



'''

df17 = pd.read_sql_query(q17, db) 
df17
```

Mountain bikes seem to be the best-selling category of bikes, with 2.7 million dollars in total revenue across all brands.

```python
q18 = '''
SELECT staffs.first_name, staffs.last_name, s.store_name, 
'$' ||
ROUND(((SUM(ort.quantity*(ort.list_price*(1-ort.discount))))/1000),2) || ' thousand' 
AS revenue
FROM staffs
LEFT JOIN orders o 
ON o.staff_id=staffs.staff_id
LEFT JOIN order_items ort 
ON ort.order_id=o.order_id
LEFT JOIN stores s 
ON s.store_id=staffs.store_id
GROUP BY staffs.staff_id 
ORDER BY 
((SUM(ort.quantity*(ort.list_price*(1-ort.discount))))) DESC;



'''

df18 = pd.read_sql_query(q18, db) 
df18
```

Marcelene Boyer, followed by Venita Daniel, has generated the most revenue across all three stores. We can create a subquery to show the most productive employee for each store.

```python
q19 = '''
SELECT store_name, first_name, last_name, MAX(revenue) as revenue FROM
(
SELECT staffs.first_name, staffs.last_name, s.store_name, 
'$' ||
ROUND(((SUM(ort.quantity*(ort.list_price*(1-ort.discount))))/1000),2) || ' thousand' 
AS revenue
FROM staffs
LEFT JOIN orders o 
ON o.staff_id=staffs.staff_id
LEFT JOIN order_items ort 
ON ort.order_id=o.order_id
LEFT JOIN stores s 
ON s.store_id=staffs.store_id
GROUP BY staffs.staff_id 
ORDER BY 
((SUM(ort.quantity*(ort.list_price*(1-ort.discount))))) DESC
)
GROUP BY store_name;



'''

df19 = pd.read_sql_query(q19, db) 
df19
```

This shows the most productive employee (revenue-generating) for each store.

# **Inventory & Operations Analysis**


```python
q21= '''

SELECT p.product_name 
AS products_out_of_stock 
FROM products p
LEFT JOIN stocks s
ON s.product_id=p.product_id
LEFT JOIN stores st
ON st.store_id=s.store_id
WHERE s.quantity=0 AND store_name='Baldwin Bikes'


'''

df21 = pd.read_sql_query(q21, db) 
df21
```

This represents the products that are out of stock for Baldwin's bikes.

By altering the store name in the WHERE clause, we can see the out-of-stock products for each store.

```python
q21a= '''
SELECT store_name, COUNT(products_out_of_stock) AS Amount_out_of_stock FROM
(
SELECT st.store_name,p.product_name 
AS products_out_of_stock 
FROM products p
LEFT JOIN stocks s
ON s.product_id=p.product_id
LEFT JOIN stores st
ON st.store_id=s.store_id
WHERE s.quantity=0
)
GROUP BY
store_name
ORDER BY
Amount_out_of_stock DESC;



'''

df21a = pd.read_sql_query(q21a, db) 
df21a
```

```python
q22= '''

SELECT p.product_name 
AS products_in_low_stock
FROM products p
LEFT JOIN stocks s
ON s.product_id=p.product_id
LEFT JOIN stores st
ON st.store_id=s.store_id
WHERE s.quantity<=5 AND store_name='Santa Cruz Bikes'


'''

df22 = pd.read_sql_query(q22, db) 
df22
```

This represents the products in low stock (less than 5) in Santa Cruz Bikes.

Again, the store name in the WHERE clause can be adjusted to show each store's data.

```python
q22a= '''
SELECT store_name, COUNT(products_in_low_stock) AS Amount_low_in_stock FROM
(
SELECT st.store_name, product_name 
AS products_in_low_stock
FROM products p
LEFT JOIN stocks s
ON s.product_id=p.product_id
LEFT JOIN stores st
ON st.store_id=s.store_id
WHERE s.quantity<=5
)
GROUP BY store_name ORDER BY Amount_low_in_stock DESC;


'''

df22a = pd.read_sql_query(q22a, db) 
df22a
```

```python
q23= '''

SELECT 
strftime('%m', o.order_date) AS month, '$' ||
ROUND(SUM(ort.quantity * ort.list_price * (1 - ort.discount)), 2) AS revenue
FROM orders o
LEFT JOIN order_items ort 
ON ort.order_id = o.order_id
WHERE strftime('%Y', o.order_date) = '2016'
GROUP BY strftime('%m', o.order_date) 
ORDER BY
ROUND(SUM(ort.quantity * ort.list_price * (1 - ort.discount)), 2) DESC;


'''

df23 = pd.read_sql_query(q23, db) 
df23
```

This shows the total revenue for all bike stores grouped by every month in 2016. For all stores cumulative, the best month was September. Month can be adjusted with the WHERE clause.

To show each store, we join the 'stores' table and add a WHERE clause to define the specific store we want data.

```python
q24= '''

SELECT 
strftime('%Y', o.order_date) AS year, '$' ||
ROUND(SUM(ort.quantity * ort.list_price * (1 - ort.discount)), 2) AS revenue
FROM orders o
LEFT JOIN order_items ort 
ON ort.order_id = o.order_id
GROUP BY strftime('%Y', o.order_date) 
ORDER BY year DESC;


'''

df24 = pd.read_sql_query(q24, db) 
df24
```

We find that the best-performing year for all three stores cumulative was 2017, with 3.4 million combined from all stores.

# **Marketing Analytics**

```python


q25= '''


SELECT cu.first_name,cu.last_name
FROM customers cu 
LEFT JOIN orders o 
ON o.customer_id=cu.customer_id
LEFT JOIN stores s
ON s.store_id=o.store_id
WHERE s.store_name='Baldwin Bikes'
GROUP BY cu.customer_id 
HAVING COUNT(DISTINCT o.order_id)>=3



'''

df25 = pd.read_sql_query(q25, db) 
df25
```

This shows loyal customers (Ordering 3 or more times) from Baldwin Bikes. 

The WHERE clause in the query can be used to change the store name and find loyal customers from each store.

```python
q26='''
SELECT cu.first_name, cu.last_name,s.state, s.zip_code, '$' ||
ROUND(SUM(ort.quantity*(ort.list_price-(ort.discount*ort.list_price))),2) AS total_spent
FROM customers cu 
LEFT JOIN orders o 
ON o.customer_id=cu.customer_id
LEFT JOIN order_items ort 
ON ort.order_id=o.order_id 
LEFT JOIN stores s 
ON s.store_id=o.store_id
WHERE s.store_name='Santa Cruz Bikes'
GROUP BY cu.customer_id
ORDER BY ROUND(SUM(ort.quantity*(ort.list_price-(ort.discount*ort.list_price))),2) 
DESC LIMIT 10;
'''
df26=pd.read_sql_query(q26,db)
df26
```

This shows the top-spending customers for Santa Cruz Bikes. We can look at each store's data by changing the store in the WHERE clause.
