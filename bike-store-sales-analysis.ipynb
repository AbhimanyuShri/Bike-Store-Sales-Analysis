# **Which regions are most profitable per order, and should we target marketing/expansion there?**

```python
q11 = '''

SELECT 
    str.state,'$' || 
    CAST(ROUND(SUM(ort.quantity*(ort.list_price*(1-ort.discount))) / COUNT(DISTINCT ort.order_id),2) AS int)
AS avg_revenue_per_order
FROM stores str 
LEFT JOIN orders o 
    ON o.store_id=str.store_id 
LEFT JOIN order_items ort
    ON ort.order_id=o.order_id 
    GROUP BY 
        str.state 
    ORDER BY 
        avg_revenue_per_order DESC;
    
'''

df11 = pd.read_sql_query(q11, db) 
df11
```

Identifies the average revenue a customer would make per order (total revenue/total orders) grouped by state.
As we can see, Texans tend to spend the most per order, so online services could be informed to focus on Texas markets.

# **Are we successfully retaining customers overall and at each store?**

```python
q12 = '''

SELECT status, COUNT(status) AS 'Amount'  FROM
    (
    SELECT cu.first_name,cu.last_name,
        CASE WHEN COUNT(DISTINCT o.order_id)>1 THEN 'returning'
        ELSE 'new'
    END AS 'status'
    FROM customers cu 
    LEFT JOIN orders o
        ON o.customer_id=cu.customer_id
    GROUP BY 
        cu.customer_id
    )
GROUP BY 
    status;
    
'''

df12 = pd.read_sql_query(q12, db) 
df12
```

Identifies how many customers are new and how many are returning via a subquery for all three stores. 

```python
qx = '''

SELECT 
  store_name,
  COUNT(CASE WHEN status = 'returning' THEN 1 END) AS returning_customers,
  COUNT(CASE WHEN status = 'new' THEN 1 END) AS new_customers,
  ROUND(
    100.0 * COUNT(CASE WHEN status = 'returning' THEN 1 END) 
    / COUNT(*), 
    2
  ) || '%' AS returning_percentage
FROM (
  SELECT 
    cu.customer_id,
    s.store_name,
    CASE 
      WHEN COUNT(DISTINCT o.order_id) > 1 THEN 'returning'
      ELSE 'new'
    END AS status
  FROM customers cu
  LEFT JOIN orders o ON o.customer_id = cu.customer_id
  LEFT JOIN stores s ON s.store_id = o.store_id
  GROUP BY cu.customer_id, s.store_name
) customer_status_by_store
GROUP BY store_name ORDER BY 
ROUND(
    100.0 * COUNT(CASE WHEN status = 'returning' THEN 1 END) 
    / COUNT(*), 
    2
  ) DESC
  
'''

dfx = pd.read_sql_query(qx, db) 
dfx
```

Although Baldwin Bikes have had the most returning customers, Santa Cruz Bikes has the largest retention rate of customers, with 18.31% of all of its customers returning.

# **Which brands drive our revenue and deserve better placement/promotion in stores?**

```python
q15 = '''

SELECT b.brand_name, '$' ||
CAST(ROUND(((SUM(ort.quantity*(ort.list_price*(1-ort.discount))))/1000),1) AS int) || ' thousand'
AS revenue
FROM brands b 
LEFT JOIN products p 
ON p.brand_id=b.brand_id
LEFT JOIN order_items ort 
ON ort.product_id=p.product_id 
GROUP BY 
    b.brand_name 
ORDER BY 
    ROUND(((SUM(ort.quantity*(ort.list_price-(ort.discount*ort.list_price))))/1000),2) DESC;



'''

df15 = pd.read_sql_query(q15, db) 
df15
```

The Trek bike brand is doing the best in terms of generating revenue across all three stores.

# **Which staff should be rewarded or trained? Who influences the most revenue per store?**

```python
q18 = '''

SELECT staffs.first_name, staffs.last_name, staffs.email, staffs.phone, s.store_name, 
'$' ||
CAST(ROUND(((SUM(ort.quantity*(ort.list_price*(1-ort.discount))))/1000),2) AS int) || ' thousand' 
AS revenue
FROM staffs
LEFT JOIN orders o 
ON o.staff_id=staffs.staff_id
LEFT JOIN order_items ort 
ON ort.order_id=o.order_id
LEFT JOIN stores s 
ON s.store_id=staffs.store_id
GROUP BY 
    staffs.staff_id 
ORDER BY 
    ((SUM(ort.quantity*(ort.list_price*(1-ort.discount))))) DESC;



'''

df18 = pd.read_sql_query(q18, db) 
df18
```

Marcelene Boyer, followed by Venita Daniel, has generated the most revenue across all three stores. We can create a subquery to show the most productive employee for each store.

```python
q19 = '''

SELECT store_name, first_name, last_name, email, phone, MAX(revenue) as revenue FROM
    (
    SELECT staffs.first_name, staffs.last_name, staffs.email, staffs.phone, s.store_name, 
    '$' ||
    CAST(ROUND(((SUM(ort.quantity*(ort.list_price*(1-ort.discount))))/1000),2) AS int) || ' thousand' 
    AS revenue
    FROM staffs
    LEFT JOIN orders o 
    ON o.staff_id=staffs.staff_id
    LEFT JOIN order_items ort 
    ON ort.order_id=o.order_id
    LEFT JOIN stores s 
    ON s.store_id=staffs.store_id
        GROUP BY 
            staffs.staff_id 
    ORDER BY 
        ((SUM(ort.quantity*(ort.list_price*(1-ort.discount))))) DESC
    )
GROUP BY 
    store_name;



'''

df19 = pd.read_sql_query(q19, db) 
df19
```

This shows the most productive employee (revenue-generating) for each store.

# **What stores do we have to focus on currently concerning our stock?**

```python
q21= '''

SELECT p.product_name
AS products_out_of_stock, st.store_name  
FROM products p
LEFT JOIN stocks s
ON s.product_id=p.product_id
LEFT JOIN stores st
ON st.store_id=s.store_id
WHERE 
    s.quantity=0 
AND 
    store_name='Baldwin Bikes'
    --Can change store name here--


'''

df21 = pd.read_sql_query(q21, db) 
df21
```

This represents the products that are out of stock for Baldwin's bikes.

By altering the store name in the WHERE clause, we can see the out-of-stock products for each store.

```python
q21a= '''

SELECT store_name, state, COUNT(products_out_of_stock) AS Amount_out_of_stock FROM
    (
    SELECT st.store_name, st.state, p.product_name 
    AS products_out_of_stock 
    FROM products p
    LEFT JOIN stocks s
    ON s.product_id=p.product_id
    LEFT JOIN stores st
    ON st.store_id=s.store_id
    WHERE 
        s.quantity=0
    )
GROUP BY
    store_name
ORDER BY
    Amount_out_of_stock DESC;



'''

df21a = pd.read_sql_query(q21a, db) 
df21a
```

```python
q22= '''

SELECT p.product_name 
AS products_in_low_stock, st.store_name
FROM products p
LEFT JOIN stocks s
ON s.product_id=p.product_id
LEFT JOIN stores st
ON st.store_id=s.store_id
WHERE 
    s.quantity<=5 
AND 
    store_name='Santa Cruz Bikes'
    --Can change store name here--


'''

df22 = pd.read_sql_query(q22, db) 
df22
```

This represents the products in low stock (less than 5) in Santa Cruz Bikes.

Again, the store name in the WHERE clause can be adjusted to show another store's data.

```python
q22a= '''

SELECT store_name,state, COUNT(products_in_low_stock) AS Amount_low_in_stock FROM
    (
    SELECT st.store_name, st.state, product_name 
    AS products_in_low_stock
    FROM products p
    LEFT JOIN stocks s
    ON s.product_id=p.product_id
    LEFT JOIN stores st
    ON st.store_id=s.store_id
    WHERE 
        s.quantity<=5
    )
GROUP BY 
    store_name 
ORDER BY 
    Amount_low_in_stock DESC;


'''

df22a = pd.read_sql_query(q22a, db) 
df22a
```

# **When should we run sales or launch new products? What are our peak seasons?**

```python
q23= '''

SELECT 
strftime('%m', o.order_date) AS month, strftime('%Y', o.order_date) AS year, '$' ||
CAST(ROUND(SUM(ort.quantity * ort.list_price * (1 - ort.discount))) AS int) AS revenue
FROM orders o
LEFT JOIN order_items ort 
ON ort.order_id = o.order_id
WHERE 
    strftime('%Y', o.order_date) = '2016'
    --Can adjust year here--

GROUP BY 
    strftime('%m', o.order_date) 
ORDER BY
    ROUND(SUM(ort.quantity * ort.list_price * (1 - ort.discount)), 2) DESC;


'''

df23 = pd.read_sql_query(q23, db) 
df23
```

This shows the total revenue for all bike stores grouped by every month in 2016. For all stores combined, the best month was September. The year can be adjusted with the WHERE clause.

To show each store, we join the 'stores' table and add a WHERE clause to define the specific store we want data, as shown enclosed in the dashes.

# **How many customers shop repeatedly, and should we implement or improve loyalty programs?**

```python


q25= '''


SELECT cu.first_name,cu.last_name, cu.email, cu.phone
FROM customers cu 
LEFT JOIN orders o 
ON o.customer_id=cu.customer_id
LEFT JOIN stores s
ON s.store_id=o.store_id
WHERE 
    s.store_name='Baldwin Bikes'
    --Can adjust store name here--

GROUP BY 
    cu.customer_id 
HAVING
    COUNT(DISTINCT o.order_id)>=3



'''

df25 = pd.read_sql_query(q25, db) 
df25
```

This shows loyal customers (Ordering 3 or more times) from Baldwin Bikes. 

The WHERE clause in the query can be used to change the store name and find loyal customers from each store.
